## Reading the dataset

```python
#Storing the dataset into a variable
data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
sns.set(style="darkgrid")
```

## Pairplot of Relationships

```python
# Selecting which columns to visualize from the dataset and then using those columns to implement a pairplot
sel_columns = ['Sleep Duration', 'Quality of Sleep','Physical Activity Level', 'Stress Level', 'Age', 'Blood Pressure']
# Implementing Pairplot
pairPlot = sns.pairplot(data[sel_columns])
# Adding a title
plt.suptitle("Table 1: Exploring Relationships Among Sleep and Health Variables", y = 1.02, fontsize = 16)
# Displaying graph
plt.show() 
           
```

## Relationships between Lifestyle factors and Sleep Duration

```python
# Visualization between exercise and sleep duration
#Scatter plot
sns.scatterplot(data = data, x = 'Sleep Duration', y = 'Physical Activity Level')
#Line of best fit
sns.regplot(data = data, x = 'Sleep Duration', y = 'Physical Activity Level', scatter = False)
plt.title('Table 2: Relationship between Exercise and Sleep Duration')
plt.xlabel('Sleep Duration (hrs)', fontsize = 15)
plt.ylabel('Exercise (minutes per day)')
```

```python
# Visualization between stress level and sleep duration
#Scatter plot
sns.scatterplot(data = data, x = 'Sleep Duration', y = 'Stress Level')
#Line of best fit
sns.regplot(data = data, x = 'Sleep Duration', y = 'Stress Level', scatter = False)
plt.title('Table 3: Relationship between Stress Levels and Sleep Duration')
plt.xlabel('Sleep Duration (hrs)', fontsize = 15)
plt.ylabel('Stress Level (Scale 1-10)')
```

```python
# Visualization between Age and sleep duration
# Scatter plot
sns.scatterplot(data = data, x = 'Sleep Duration', y = 'Age')
#Line of best fit
sns.regplot(data = data, x = 'Sleep Duration', y = 'Age', scatter = False)
plt.title('Table 4: Relationship between Age and Sleep Duration')
plt.xlabel('Sleep Duration (hrs)', fontsize = 15)
plt.ylabel('Age')
```
